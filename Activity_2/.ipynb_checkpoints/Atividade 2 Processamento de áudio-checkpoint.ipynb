1) Criar uma função para gerar um tom senoidal.

Parâmetros de entrada: frequência (Hz), duração (s), amplitude (0 a 1) e frequência de amostragem (Hz).

Verfique se a frequência da saída está correta medindo um período do sinal.

```python
def tonegen(freq, duration, A, fs):
    if A < 0: 
        A = 0
    elif A > 1:
        A = 1
    t = np.linspace(0., duration, int(fs * duration))
    x = A*np.sin(freq * 2. * np.pi * t)
    
    return x

    
```

```python
s440 = tonegen(80, .1, 0.8, 8000)

t = np.linspace(0.,  .1, int(8000 *  .1))
plt.stem(t, s440,'--', )
#plt.figure(2)
#plt.plot(s440,'bo', s440, 'k')
plt.show()
```

2) Criar uma função que receba um vetor com uma sequência de dígitos e gere tons DTMF correspondentes. Os parâmetros de entrada da função são: sequência de dígitos, duração em segundos de cada dígito, intervalo em segundos entre cada dígito e amplitude (0 a 1). O sinal de saída deve ter frequência de amostragem de 8 kHz.

Para verificar se o sinal está correto, teste a saída usando um decodificador DTMF, como os listados:
- online: DialABC http://dialabc.com/sound/
- Android DTMF Transceiver https://play.google.com/store/
- Android DTMF Decoder https://play.google.com/store/
- OcenAudio, em Análise/Detectar tons DTMF

```python
def dtmfgen(seq, t, ts,A):
    smpl = 8000
    DTMF_C=[1209, 1336, 1477, 1633]
    DTMF_L=[697, 770, 852, 941]
    DTMF_T=[  1, 2,   3, 'A',
              4, 5,   6, 'B', 
              7, 8,   9, 'C', 
            '*', 0, '#', 'D']

    n=len(seq)
    
    I = int(DTMF_T.index(seq[0])/4)
    J = DTMF_T.index(seq[0])%4
    
    
    fg=tonegen(DTMF_L[I],t,A,smpl)/2 + tonegen(DTMF_C[J],t,A,smpl)/2
    zeros = tonegen(10,ts,0,smpl)
    s =  np.append(fg,zeros)
    
    N = 1
    while N < n :
        I = int(DTMF_T.index(seq[N])/4)
        J = DTMF_T.index(seq[N])%4
    
    
        fg=tonegen(DTMF_L[I],t,A,smpl)/2 + tonegen(DTMF_C[J],t,A,smpl)/2
        zeros = tonegen(10,ts,0,smpl)
        fg =  np.append(fg,zeros)
        s =  np.append(s,fg)
        N = N + 1
 
    display(Audio(s, rate=smpl, autoplay=True)) # para mostrar o widget
    return s

```

3) Criar uma função para remover uma em cada duas amostras de um sinal, criando um novo sinal que tem somente as amostras de índice ímpar do sinal original.

```python
def halfsp(x):
    s=x[::2]
    return s
```

```python
x=np.array([7, 3, 9, 1, 0, 4])
half_x = halfsp(x)
print(half_x)
#resultado esperado: [7, 9, 0]
```

4) Criar uma função para criar um novo sinal que insere uma nova amostra entre cada duas amostras. Essa nova amostra é a média das duas amostras vizinhas.

```python
def doublesp(x):
    tam = len(x)
    s=np.zeros(2*tam-1)
    
    s[::2] = x[::]

    s[1:-1:2] = (x[0:-1:] + x[1::])/2

    
    return s
```

```python
x=np.array([7, 3, 9, 1, 0, 4])
double_x = doublesp(x)
print(double_x)
#[7 5 3 6 9 5 1 0.5 0 2 4]
```

5) Gere um tom de 220 Hz. Aplique a função `halfsp`. Escute o resultado antes e depois da aplicação. O que ocorre na frequência e no tempo? Mostre graficamente o resultado na frequência através da transformada de Fourier (implementada pela função `numpy.fft`).

```python
smpl = 8000
total_time = 0.1

s220 = tonegen(220, total_time, 0.8, smpl)
half_s220 = halfsp(s220)

display(Audio(s220, rate=smpl, autoplay=True)) # para mostrar o widget
display(Audio(half_s220, rate=smpl, autoplay=True)) # para mostrar o widget
```

Eliminando metade das amostras do vetor o sinal perde resolução e dobra a frequência. De 220Hz vai para 440Hz, isso pode ser percebido ouvindo que o sinal ficou mais agudo e fica claro nos gráficos de fft de cada sinal.


```python
n = len(s220)
Fs = smpl
Ts = total_time/Fs; # sampling interval
t = np.arange(0,1,Ts) # time vector

k = np.arange(n)
T = n/Fs
frq = k/T # two sides frequency range
frq = frq[range(int(n/2))] # one side frequency range

Y = np.fft.fft(s220)/n # fft computing and normalization
Y = Y[range(int(n/2))]


fig, ax = plt.subplots(1, 1)
ax.plot(frq,abs(Y),'r') # plotting the spectrum
ax.set_title('S220 fft')
ax.set_xlabel('Freq (Hz)')
ax.set_ylabel('| s220(A)|')
plt.show()
```

```python
n = len(half_s220)
Fs = smpl
Ts = total_time/2/Fs; # sampling interval
t = np.arange(0,1,Ts) # time vector

k = np.arange(n)
T = n/Fs
frq = k/T # two sides frequency range
frq = frq[range(int(n/2))] # one side frequency range

Y = np.fft.fft(half_s220)/n # fft computing and normalization
Y = Y[range(int(n/2))]


fig, ax = plt.subplots(1, 1)
ax.plot(frq,abs(Y),'r') # plotting the spectrum
ax.set_title('S440 = halsp(s220) fft')
ax.set_xlabel('Freq (Hz)')
ax.set_ylabel('| half_s220(A)|')
plt.show()
```
